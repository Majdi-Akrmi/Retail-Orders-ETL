### Read data from the file

```python
import pandas as pd

df = pd.read_csv('orders.csv')
df.head(25)
```

### Transforming Data

```python
# Handle the null value

df = pd.read_csv('orders.csv', na_values=['Not Available', 'unknown'])
df['Ship Mode'].unique()
```

```python
# Rename columns names

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns
df.head(5)
```

```python
# Derive new columns discount, sale_price and profit

df['discount'] = df['list_price'] * df['discount_percent'] * .01
df['sale_price'] = df['list_price'] - df['discount']
df['profit'] = df['sale_price'] - df['cost_price']
```

```python
df
```

```python
# Convert order_date from object data type to datetime
df.dtypes
df['order_date'] = pd.to_datetime(df['order_date'], format = "%Y-%m-%d")
```

```python
df.dtypes
```

```python
# Remove cost_price, list_price and discount_percent columns

df.drop(columns = ['cost_price', 'list_price', 'discount_percent'], inplace = True)
```

```python
# Connect to PostgreSQL database
from sqlalchemy import create_engine
 
engine = create_engine('postgresql+psycopg2://postgres:postgres@localhost/Order_db')
conn = engine.connect()
```
